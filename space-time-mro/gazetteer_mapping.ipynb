```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
df = pd.read_csv("gazetteerToMap.csv", header=0) 
```

```python
df.head()
```

```python
df.plot(kind="scatter", x="long", y="lat", alpha=0.4)     
```

```python
geometry = [Point(xy) for xy in zip(df['long'], df['lat'])]
gdf = GeoDataFrame(df, geometry=geometry)   

#Activate below for interactive mapping

#%matplotlib notebook 

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(20, 12)), marker='o', color='red', markersize=1, alpha=0.4);
```
